```python
from ebaysdk.finding import Connection as Finding
import json

#Fetching data from ebay API
api = Finding(domain='svcs.ebay.com', appid="PRUDHVIS-Assigmen-PRD-5393cab7d-2fed1ef4", config_file=None)
response = api.execute('findItemsByKeywords', {'keywords': 'bose headphones'})
output = response.dict()
output["searchResult"]["item"]
```

```python
#Defing lists to store data
product_list =[]
shipping_type=[]
payment_method=[]
selling_state =[]


for i in (output["searchResult"]["item"]):
    #Product name
    title = i["title"]
    
    #Payment Method
    paymentmethod = i["paymentMethod"]
    
    #Shipping Type
    shippingtype=i['shippingInfo']["shippingType"]
    
    #Selling Status
    sellingstate=i["sellingStatus"]["sellingState"]
    
    product_list.append(title)
    shipping_type.append(shippingtype)
    payment_method=paymentmethod
    
    selling_state.append(sellingstate)
    
   
```

```python
import pandas as ps

#Adding lists to pandas dataframe
Product_data  = ps.DataFrame({"PRODUCT_NAME":product_list,
                              "SHIPPNG_TYPE":shipping_type,
                             "PAYMENT_METHOD":payment_method,
                             "SELLING_STATUS":selling_state})
```

```python
#Remocing duplicates if any
Product_data.drop_duplicates
```
